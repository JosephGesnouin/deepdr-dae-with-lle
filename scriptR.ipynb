```r
library(readr)
library(mclust)
library(aricode)
```

## Lambda 0.2, 10, 10

```r
bottle <- read_csv("~/Developpement/cours/DeepReduc/bottleneck.csv")
labels <- read_csv("~/Developpement/cours/DeepReduc/labels.csv")
nmi=c()
ari=c()
for(i in 1:3){
  clustering=Mclust(bottle,G=10)
  nmi=cbind(nmi,NMI(clustering$classification,as.matrix(labels)[,1]))
  ari=cbind(ari,ARI(clustering$classification,as.matrix(labels)[,1]))
}
NMIPCA=as.vector(nmi);mean(NMIPCA)
ARIPCA=as.vector(ari);mean(ARIPCA)
boxplot(NMIPCA)
boxplot(ARIPCA)
```
